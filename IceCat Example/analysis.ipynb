```python
import pandas as pd
```

## Data Profiling

```python

print('Data Preview:')
complete_icecat_data = pd.read_csv('icecat_office_products_100.csv', dtype=str)
complete_icecat_data.head()
```

```python

print('Number of products per individual category (top 10):')
complete_icecat_grpdby_category = complete_icecat_data.groupby(by='category_name').count() \
                                                        .sort_values(by=['id'], ascending=False) \
                                                        .reset_index()

complete_icecat_grpdby_category.iloc[:, 0:2].rename(columns={'id': 'count'}).head(10)
```

```python

rank = 3
category_col_count = complete_icecat_grpdby_category.iloc[rank,:]
print(f"Columns with full data for {category_col_count['category_name']}:")
category_col_count[category_col_count == category_col_count['id']]
```
